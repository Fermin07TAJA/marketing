```python
import numpy as np
import matplotlib.pyplot as plt

years = np.array([1,2,3,4,5])
revenue = np.array([1.0, 1.5, 4.0, 7.5, 12.0])
margins = np.array([0.05,0.10,0.18,0.28,0.40])
profit = revenue*margins

fig, ax = plt.subplots(figsize=(8,5))
fig.patch.set_facecolor("#D0E1BE")
ax.set_facecolor("#D0E1BE")

ax.plot(years, revenue, color="#2a6f37", linewidth=3, label="Revenue (M$)")
ax.plot(years, profit, color="black", linewidth=3, label="Profit (M$)")

ax.set_xticks(years)
ax.set_xlabel("Year", color="black")
ax.set_ylabel("Million USD", color="black")
ax.set_title("Projected Revenue and Profit", color="black")

ax.tick_params(colors="black")
ax.legend()

plt.savefig("projection.png", dpi=300, bbox_inches="tight")
plt.show()

```
